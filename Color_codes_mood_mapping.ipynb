# Color Codes for emotions
The colors are determined considering the visualization of the literature for emotions

```python
import matplotlib.pyplot as plt
# Creating a simple plot to display the colors for each emotion

color_codes = {
    'Joy': '#FFFF00',  # Yellow bright
    'Sadness': '#6495ED',  # Blue
    'Anger': '#FF0000',  # Slightly light red
    'Fear': '#008000',  # Green
    'Love': '#F06292',  # Pinkish orange
    'Surprise': '#9B59B6'  # Between blue and green
}

fig, ax = plt.subplots(1, len(color_codes), figsize=(12, 2))

for i, (emotion, color) in enumerate(color_codes.items()):
    ax[i].set_facecolor(color)
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_title(emotion)

plt.tight_layout()
plt.show()
```

## Coloring Moods
We determined colors considering the emotions moods consist.

**Relaxed**

High Scores: Love

Low Scores: Anger

A state of tranquility and comfort, characterized by feelings of affection and calmness, with an absence of aggression or irritability.

**Cheerful**

High Scores: Joy

Low Scores: Sadness, Anger

Represents a happy, content state, where one experiences joy and optimism, while feelings of sorrow and aggression are minimal.

**Nostalgic**

High Scores: Sadness

Low Scores: Surprise, Joy

A mood of longing or wistfulness, often for the past, marked by a touch of sadness and a lack of surprise or joy.

**Adventurous**

High Scores: Surprise

Low Scores: Fear, Sadness

Characterized by a desire for new experiences and excitement, with minimal feelings of fear or sorrow.

**Contemplative**

High Scores: Love

Low Scores: Joy

A reflective and introspective state, often filled with deep emotions like affection, but with a lower presence of outright happiness or elation.

**Energetic**

High Scores: Joy

Low Scores: Fear

```python
mood_colors = {
    'Cheerful': '#FFEB3B',  # Yellow
    'Nostalgic': '#009688',  # Softer blue
    'Adventurous': '#800080',  # Purple
    'Contemplative': '#8B0000',  # Deep red
    'Energetic': '#FF6F00',  # Bright yellow
    'Relaxed': '#9ACD32',  # Bright yellow
}

fig, ax = plt.subplots(1, len(mood_colors), figsize=(12, 2))

for i, (mood, color) in enumerate(mood_colors.items()):
    ax[i].set_facecolor(color)
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_title(mood)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define the hexagon vertices and colors for each mood/emotion
mood_colors = {
    'Cheerful': '#FFEB3B',  # Yellow
    'Nostalgic': '#009688',  # Softer blue
    'Adventurous': '#800080',  # Purple
    'Contemplative': '#8B0000',  # Deep red
    'Energetic': '#FF6F00',  # Bright yellow
    'Relaxed': '#9ACD32',  # Bright green
}

# Define the coordinates for the hexagon vertices
hexagon_vertices = np.array([
    [1, 0],  # Right
    [0.5, np.sqrt(3) / 2],  # Top-right
    [-0.5, np.sqrt(3) / 2],  # Top-left
    [-1, 0],  # Left
    [-0.5, -np.sqrt(3) / 2],  # Bottom-left
    [0.5, -np.sqrt(3) / 2],  # Bottom-right
    [1, 0]  # Closing the hexagon to the right vertex
])

# Calculate the center point of the hexagon for mood mapping
center_point = np.mean(hexagon_vertices[:-1], axis=0)

# Plot the hexagon and map the moods
fig, ax = plt.subplots(figsize=(10, 8))
ax.add_patch(mpatches.Polygon(hexagon_vertices, closed=True, fill=None, edgecolor='k'))

# Plot the moods on the corners of the hexagon
for i, (mood, color) in enumerate(mood_colors.items()):
    ax.scatter(*hexagon_vertices[i], color=color, s=100, label=mood)

# Map the moods to inner space (placeholder coordinates)
# You would replace these with the actual values from your mood incentives analysis
for mood, color in mood_colors.items():
    ax.scatter(*center_point, color=color, s=100)

ax.set_title('Mood Mapping on a Hexagon')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.axis('equal')  # Equal aspect ratio
plt.axis('off')  # Hide the axis
plt.show()
```

```python
# Define the hexagon vertices and colors for each emotion
emotion_colors = {
    'Joy': '#FFFF00',  # Yellow bright
    'Sadness': '#6495ED',  # Blue
    'Anger': '#FF0000',  # Slightly light red
    'Fear': '#008000',  # Green
    'Love': '#F06292',  # Pinkish orange
    'Surprise': '#9B59B6'  # Between blue and green
}

# Define the coordinates for the hexagon vertices
hexagon_vertices = np.array([
    [0, 1],  # Top
    [np.sqrt(3)/2, 0.5],  # Top-right
    [np.sqrt(3)/2, -0.5],  # Bottom-right
    [0, -1],  # Bottom
    [-np.sqrt(3)/2, -0.5],  # Bottom-left
    [-np.sqrt(3)/2, 0.5],  # Top-left
    [0, 1]  # Back to top to close hexagon
])

# Define the coordinates for the moods based on their emotion intensity
# These are placeholders and should be replaced with your actual data
mood_coordinates = {
    'Cheerful': [0.5, 0.8],
    'Nostalgic': [-0.4, 0.6],
    'Adventurous': [0.6, -0.1],
    'Contemplative': [-0.5, -0.3],
    'Energetic': [0.3, -0.7],
    'Relaxed': [-0.1, -0.8]
}

# Plot the hexagon and map the emotions to the vertices
fig, ax = plt.subplots(figsize=(8, 8))
ax.add_patch(mpatches.Polygon(hexagon_vertices, closed=True, fill=None, edgecolor='k'))

# Plot the emotions on the corners of the hexagon
for i, (emotion, color) in enumerate(emotion_colors.items()):
    ax.scatter(*hexagon_vertices[i], color=color, s=100, label=emotion)

# Plot the moods inside the hexagon according to their intensity
for mood, coordinates in mood_coordinates.items():
    ax.scatter(*coordinates, color=mood_colors[mood], s=100, label=mood, edgecolors='k')

# Set the axes limits and add a legend
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_title('Mood and Emotion Mapping on a Hexagon')
plt.axis('equal')  # Equal aspect ratio
plt.axis('off')  # Hide the axis
plt.show()
```
