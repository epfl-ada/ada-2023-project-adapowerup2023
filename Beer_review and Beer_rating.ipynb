```python
BA_ratings = pd.read_csv('.\\Filtered\\BA_ratings.csv')
BA_reviews = pd.read_csv('.\\Filtered\\BA_reviews.csv')
MBD_ratings = pd.read_csv('.\\Filtered\\MBD_ratings.csv')#compared dataset
MBD_ratings_ba = pd.read_csv('.\\Filtered\\MBD_ratings_ba.csv')
MBD_ratings_rb = pd.read_csv('.\\Filtered\\MBD_ratings_rb.csv')
MBD_ratings_with_text_ba = pd.read_csv('.\\Filtered\\MBD_ratings_with_text_ba.csv')
MBD_ratings_with_text_rb = pd.read_csv('.\\Filtered\\MBD_ratings_with_text_rb.csv')
RB_ratings = pd.read_csv('.\\Filtered\\RB_ratings.csv')

merged_dataset = pd.concat([BA_ratings, BA_reviews], axis=0)
merged_dataset = pd.concat([MBD_ratings_ba, merged_dataset], axis=0)
merged_dataset = pd.concat([MBD_ratings_rb, merged_dataset], axis=0)
merged_dataset = pd.concat([MBD_ratings_with_text_ba, merged_dataset], axis=0)
merged_dataset = pd.concat([MBD_ratings_with_text_rb, merged_dataset], axis=0)
merged_dataset = pd.concat([RB_ratings, merged_dataset], axis=0)

merged_dataset.to_csv(".\\Filtered\\merged_dataset.csv", index=False)
```

```python
merged_dataset = pd.read_csv(".\\Filtered\\merged_dataset.csv", sep = ",", header = 0, low_memory=False)
display(merged_dataset)
```

```python
#data cleaning, check if any data has been droped
merged_dataset_filtered = merged_dataset.copy().drop_duplicates()
print(len(merged_dataset_filtered))
```

```python
merged_dataset_counts = merged_dataset_filtered['beer_id'].value_counts()
merged_dataset_counts = pd.DataFrame(data = merged_dataset_counts)
print(merged_dataset_counts.describe())
```

```python
# calculate 'beer_id' records
beer_id_counts = merged_dataset_filtered['beer_id'].value_counts()

# Find 'beer_id' with more than 15 reviews
popular_beer_ids = beer_id_counts[beer_id_counts >= beer_id_counts.quantile(0.75)].index

# Filter 'beer_id' with more than 15 reviews
filtered_dataset = merged_dataset_filtered[merged_dataset_filtered['beer_id'].isin(popular_beer_ids)]

print(filtered_dataset)
```

```python
filtered_dataset_counts = filtered_dataset['beer_id'].value_counts()
filtered_dataset_counts = pd.DataFrame(data = filtered_dataset_counts)
print(filtered_dataset_counts.describe())
```
