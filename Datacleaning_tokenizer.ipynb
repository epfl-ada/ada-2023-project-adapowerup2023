```python
from transformers import BertTokenizer
import pandas as pd
import torch
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the filtered data from CSV files
ba_data = pd.read_csv('/content/drive/MyDrive/FDH Project/filtered_data_ba.csv')
rb_data = pd.read_csv('/content/drive/MyDrive/FDH Project/filtered_data_rb.csv')
```

```python
# Initialize the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the text data
tokenized_data_ba = ba_data['text'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, padding='max_length', max_length=512))
tokenized_data_rb = rb_data['text'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, padding='max_length', max_length=512))
```

```python
# Convert tokenized data to tensors
tokenized_tensors_ba = torch.tensor(tokenized_data_ba.tolist())
tokenized_tensors_rb = torch.tensor(tokenized_data_rb.tolist())

# Save the tokenized tensors to Google Drive
torch.save(tokenized_tensors_ba, '/content/drive/My Drive/FDH Project/tokenized_data_ba.pt')
torch.save(tokenized_tensors_rb, '/content/drive/My Drive/FDH Project/tokenized_data_rb.pt')
```
